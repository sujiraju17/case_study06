# 4. Fill in all the missing values present in all the columns in the dataset

```python
data.isna().sum()
```

```python
#find median of age
data['Age'].median()
```

```python
#filling age with median
data['Age'].fillna(round(data['Age'].median()), inplace = True)
```

# 5. Check and handle outliers in at least 3 columns in the dataset

# age

```python
sns.boxplot(data1['Age'])
```

```python
sns.distplot(data1['Age'])
```

```python
# Identifying the outliers
Q1 = np.percentile(data1['Age'],25,interpolation='midpoint')
Q3 = np.percentile(data1['Age'],75,interpolation='midpoint')
IQR = Q3-Q1
low_limit = Q1-1.5*IQR     
up_limit = Q3 + 1.5*IQR
outlier = []
for x in data1['Age']:
    if (x > up_limit) or (x < low_limit):
        outlier.append(x)       
```

```python
outlier_data1 = data1['Age'] > up_limit
data1.loc[outlier_data1]
```

```python
#replacing outliers with median
median=data1['Age'].median()
data1['Age'] = np.where(data1["Age"] >up_limit, median,data1['Age'])
data1['Age'] = np.where(data1["Age"] <low_limit, median,data1['Age'])
data1.head()
```

# fare

```python
sns.distplot(data1['Fare'])
```

```python
# Identifying the outliers
Q1 = np.percentile(data1['Fare'],25,interpolation='midpoint')
Q3 = np.percentile(data1['Fare'],75,interpolation='midpoint')
IQR = Q3-Q1
low_limit = Q1-1.5*IQR     
up_limit = Q3 + 1.5*IQR
outlier = []
for x in data1['Fare']:
    if (x > up_limit) or (x < low_limit):
        outlier.append(x)
outlier_data2 = data1['Fare'] > up_limit
data1.loc[outlier_data2]
```

```python
# Applying StandardScaler to handle the outliers 
from sklearn.preprocessing import StandardScaler
Fare=data1['Fare']
Fare=Fare.values.reshape(-1,1)
# create a scaler object
s_scaler = StandardScaler()
# fit and transform the data
scaled_Fare = pd.DataFrame(s_scaler.fit_transform(Fare))
data1['Fare']=scaled_Fare
data1.head()
```

# 6. Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
data1[['Survived']].describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# create a scaler object
s1 = MinMaxScaler()
survived_matrix=data1['Survived'].values.reshape(-1,1)
data_norm = pd.DataFrame(s1.fit_transform(survived_matrix))
data1['Survived']=data_norm
data1.head()
```

```python
# Define the dependent and independent variables 
y=data['Survived']
X=data.drop(['Survived', 'Name', 'Sex', 'Ticket','Cabin', 'Embarked'], axis=1)
```

```python
# Import train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
#linear regression model
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(X_train,y_train)
predictions=model.predict(X_test)
print(y_test)
```

```python
#Get the coefficient and intercept value of the model 
print("The model coefficient is:", model.coef_)
print("The model intercept value is:", model.intercept_)

# To evaluate the model using MSE and R squared value
from sklearn.metrics import mean_squared_error
print('MSE :',mean_squared_error(y_test,predictions))

from sklearn.metrics import r2_score
print('R squared value:',r2_score(y_test,predictions))

print("This model is a best fit , the MSE  is very small and r2_ score is small.")
```
